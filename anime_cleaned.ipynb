```python
# Replace '\n' and '\t' in 'pro_company' column
anime_movies.loc[anime_movies['pro_company'].str.contains(r'\n|\t', na=False), 'pro_company'] = \
    anime_movies['pro_company'].str.replace(r'\n|\t', '', regex=True)

anime_movies
```

```python
print(anime_movies['Language'].unique())
```

```python
translator = Translator()

# Get unique language values
unique_languages = anime_movies['Language'].dropna().unique()

# Dictionary to store translations
translation_dict = {}

# Translate each unique value
for lang in unique_languages:
    try:
        translated_text = translator.translate(lang, dest='en').text
        translation_dict[lang] = translated_text
        time.sleep(0.5)  # Prevent API rate limits
    except Exception as e:
        print(f"Error translating '{lang}': {e}")
        translation_dict[lang] = lang  # If translation fails, keep the original text

# Apply the translations back to the DataFrame
anime_movies['Language'] = anime_movies['Language'].map(translation_dict)

# Print unique translated values
print(anime_movies['Language'].unique())
```

```python
import pandas as pd
import ast

def explode_column(anime_movies, column_name, new_column_name):
    expanded_df = anime_movies[['title_id', column_name]].copy()  # Use title_id as the reference
    
    # Convert string lists into actual lists safely
    def safe_convert_list(value):
        if isinstance(value, str):  
            try:
                return ast.literal_eval(value) if value.startswith('[') else value.split(', ')
            except (SyntaxError, ValueError):
                return value.split(', ')  # If conversion fails, assume it's a comma-separated string
        elif isinstance(value, list):
            return value  # Already a list, return as is
        else:
            return [value]  # Convert single values into a list

    expanded_df[column_name] = expanded_df[column_name].apply(safe_convert_list)

    # Explode into multiple rows
    expanded_df = expanded_df.explode(column_name).dropna().reset_index(drop=True)
    
    # Rename the column
    expanded_df.rename(columns={column_name: new_column_name}, inplace=True)
    return expanded_df



# Apply the function to explode multiple columns
writers_df = explode_column(anime_movies, 'Writer', 'writer_name')
genres_df = explode_column(anime_movies, 'Genre', 'genre')
production_companies_df = explode_column(anime_movies, 'pro_company', 'company_name')
languages_df = explode_column(anime_movies, 'Language', 'languages')
country_df = explode_column(anime_movies, 'pro_country', 'country')
director_df = explode_column(anime_movies,'Director','director_name')

# Drop exploded columns from the main DataFrame
anime_movies = anime_movies.drop(columns=['Writer', 'Genre', 'pro_company', 'Language', 'pro_country','Director'])

# Display results
print("Genres DataFrame:\n", genres_df)
print("Writers DataFrame:\n", writers_df)
print("Languages DataFrame:\n", languages_df)
print("Production Companies DataFrame:\n", production_companies_df)
print("Origin Countries DataFrame:\n", country_df)
print("director DataFrame:\n",director_df)
```
